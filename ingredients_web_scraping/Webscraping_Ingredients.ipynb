# Notebook para webscrapping de ingredientes

Los ingredientes extraídos se almacenarán en un diccionario y posteriormente en un archivo json / csv.
Estas palabras se utilizarán en el reconocimiento de imágenes para filtrar las etiquetas obtenidas a través
de la API de Google Cloud Vision.

```python
from bs4 import BeautifulSoup
from termcolor import colored
import urllib
import os
import urllib.request
import json
import csv
```

## Construcción del diccionario

```python
ingredients = {}

def dig_main_page(main_url):    
    html = urllib.request.urlopen(main_url).read()
    soup = BeautifulSoup(html, 'html.parser')
    li = soup.findAll("li", {"itemprop": "item"})
    #mydivs = soup.findAll("div")
    for l in li:
        category_url = l.h3.a.get("href",None);
        category_name = l.h3.a.span.contents[0]
        print(category_name)
        print(category_url)
        dig_category(category_url, category_name)

def dig_category(category_url, category_name):
    html = urllib.request.urlopen(category_url).read()
    soup = BeautifulSoup(html, 'html.parser')
    pags = soup.findAll("span", {"class": "pagination"})
    if (len(pags) == 0):
        page_url = category_url + '/page-1'
        dig_page(page_url, category_name)
    else:
        for x in range(1, len(pags)+1):
            page_url = category_url + '/page-' + str(x)
            dig_page(page_url, category_name) 
    print('\n')
    
def dig_page(page_url, category_name):  
    html = urllib.request.urlopen(page_url).read()
    soup = BeautifulSoup(html, 'html.parser')
    li = soup.findAll("li", {"itemprop": "item"})
    #mydivs = soup.findAll("div")
    for l in li:
        ingr_name = l.h3.a.span.contents[0].lower()
        ingredients[ingr_name] = category_name.lower().replace(' ', '-')
        print(ingr_name)   
        
def webscraping():
    main_url = "https://food.ndtv.com/ingredient"
    gallery = dig_main_page(main_url)  
```

## Almacenamiento en fichero json / csv

```python
webscraping()
print(ingredients)
print(len(ingredients.keys()))
with open('ingredients.json', 'w') as fp:
    json.dump(ingredients, fp)
with open('ingredients.csv', 'w') as fp:
    for key in ingredients.keys():
        fp.write("%s,%s\n"%(key, ingredients[key]))
```
